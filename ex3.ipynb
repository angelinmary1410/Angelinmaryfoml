```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Load the data
data = pd.read_csv('suv_data-suv_data.csv')

# Select features and target, excluding User ID
X = data[['Age', 'EstimatedSalary']]
y = data['Purchased']

# Fit the logistic regression model
model = LogisticRegression().fit(X, y)
y_pred_prob = model.predict_proba(X)[:, 1]

# Plot the results
plt.figure(figsize=(12, 6))

# Plot Age vs Probability of Purchase
plt.subplot(1, 2, 1)
plt.scatter(data['Age'], y, label='Actual Purchases')
plt.scatter(data['Age'], y_pred_prob, color='red', alpha=0.5, label='Predicted Probability')
plt.xlabel('Age')
plt.ylabel('Probability of Purchase')
plt.title('Age vs Probability')
plt.legend()

# Plot Estimated Salary vs Probability of Purchase
plt.subplot(1, 2, 2)
plt.scatter(data['EstimatedSalary'], y, label='Actual Purchases')
plt.scatter(data['EstimatedSalary'], y_pred_prob, color='blue', alpha=0.5, label='Predicted Probability')
plt.xlabel('Estimated Salary')
plt.ylabel('Probability of Purchase')
plt.title('Estimated Salary vs Probability')
plt.legend()

plt.tight_layout()
plt.show()

# Print the model parameters
print(f"Intercept: {model.intercept_[0]}, Coefficients: {model.coef_[0]}")
```
